```python
def bubble_sort(arr):
    # Check if list is empty
    if not arr:
        print("Error: The list is empty.")
        return
    # Check for non-numeric values
    for item in arr:
        if not isinstance(item, (int, float)):
            print("Error: List contains non-numeric values.")
            return
    # Bubble sort algorithm
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    print("Sorted list:", arr)

# ‚úÖ Test cases
bubble_sort([5, 2, 9, 1, 5])     # Valid input
bubble_sort([])                  # Empty list
bubble_sort([3, 'a', 7])         # Non-numeric value
```

```python
# File Handling in Python with Error Handling

def read_and_write_file(input_file, output_file):
    try:
        # Read file
        with open(input_file, 'r') as f:
            data = f.read()

        # Process data (convert to uppercase)
        processed_data = data.upper()

        # Write processed data to output file
        with open(output_file, 'w') as f:
            f.write(processed_data)

        print("‚úÖ File processed and saved successfully!")

    except FileNotFoundError:
        print("‚ùå Error: Input file not found.")
    except PermissionError:
        print("‚ùå Error: Permission denied.")
    except Exception as e:
        print("‚ùå An unexpected error occurred:", e)


# Test cases
print("Case 1: Valid input file")
read_and_write_file('input.txt', 'output.txt')

print("\nCase 2: Missing file")
read_and_write_file('missing.txt', 'output.txt')
```
